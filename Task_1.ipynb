```python
from cryptography.fernet import Fernet

key =Fernet.generate_key()

cipher = Fernet(key)

name = input("Enter Your Name")
gmail = input("Enter your Gmail")

#records = name+"|"+gmail

encripted_data1 = cipher.encrypt(name.encode())
encripted_data2 = cipher.encrypt(gmail.encode())
print("Key 1->",encripted_data1)
print("Key 2->",encripted_data2)

decripted_data1 = cipher.decrypt(encripted_data1).decode()
decripted_data2 = cipher.decrypt(encripted_data2).decode()
print(decripted_data1)
print(decripted_data2)
```
